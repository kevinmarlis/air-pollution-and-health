```python
import pandas as pd
import datetime
```

```python
base_url = 'https://s3-us-west-1.amazonaws.com//files.airnowtech.org/airnow/2020/'
file = '/daily_data_v2.dat'
dates = pd.date_range(start="2020-01-01",end='today')
dates = [date.strftime("%Y%m%d") for date in dates]
urls = [f'{base_url}{date}{file}' for date in dates]
```

```python
cols = ['Valid date', 'AQSID', 'Sitename', 'Parameter name', 'Reporting units', 'Value', 'Averaging period', 'Data Source', '8-hr or 24-hr AQI Value', '8-hr or 24-hr AQI Category', 'Latitude', 'Longitude', 'Full AQSID with 3 digit country code prefix']
df = pd.read_csv(url, sep="|", header=None, names=cols)
```
