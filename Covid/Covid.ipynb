```python
covid_df = pd.read_csv('https://raw.githubusercontent.com/datadesk/california-coronavirus-data/master/latimes-place-totals.csv')
covid_df = covid_df[covid_df['county'] == 'Los Angeles'].reset_index(drop=True)
covid_df
```

```python
dates = covid_df['date'].unique()
```

```python
def average_readings(df, column_name):
    df = df[df['Date'] > '03/15/2020']
    dates = list(df['Date'].unique())
    dates.sort()
    covid_format_dates = [f'{date[6:10]}-{date[:2]}-{date[3:5]}' for date in dates if date > '03/15/2020' and date < '07/01/2020']
    daily_means = [df[df['Date'] == date][column_name].mean() for date in dates]
    dates_and_means = dict(zip(covid_format_dates, daily_means))
    return dates_and_means
```

```python
df = pd.read_csv('2020-PM25.csv')
```

```python
df_filtered = df[df['AQS_PARAMETER_DESC'] == 'PM2.5 - Local Conditions']
df_filtered
```

```python
pollutants = {}
pollutants['Daily Mean PM2.5 Concentration'] = average_readings(df_filtered, 'Daily Mean PM2.5 Concentration')
pollutants['Daily Mean PM10 Concentration'] = average_readings(pd.read_csv('2020-PM10.csv'), 'Daily Mean PM10 Concentration')
```
