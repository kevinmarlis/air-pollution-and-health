```python
import pandas as pd
import numpy as np
import missingno as msno
import seaborn as sns
%matplotlib inline
pd.set_option("display.max.columns", None)
```

```python
df = pd.read_csv('500_Cities__Census_Tract-level_Data__GIS_Friendly_Format___2019_release.csv')
```

```python
print(df.shape)
df.head()
```

```python
# Create list of TractFIPS for LA County
la_county_fips = [fip for fip in df['TractFIPS'] if str(fip)[:4] == '6037']
len(la_county_fips)
```

```python
# Filter df to only consist of LA County census tracts
la_county_df = df[df['TractFIPS'].isin(la_county_fips)]
```

```python
print(la_county_df.shape)
la_county_df
```

```python
cols = ['TractFIPS', 'CANCER_CrudePrev', 'CASTHMA_CrudePrev', 'CHD_CrudePrev', 'COPD_CrudePrev', ]
```

```python
CDC_df = la_county_df[cols]
```
