## PurpleAir and CDC 500 Cities

```python
import pandas as pd
from sodapy import Socrata
```

```python
cdc_domain = 'chronicdata.cdc.gov'
cdc_identifier = 'k86t-wghb'
```

```python
client =  Socrata(cdc_domain, None)
```

```python
results = client.get(cdc_identifier, where="tractfips LIKE '06037%'", limit=2000)
```

```python
cdc_df = pd.DataFrame.from_dict(results)
cdc_df
```

```python
cols_to_keep = ['tractfips', 'geolocation', 'bphigh_crudeprev', 'cancer_crudeprev', 'casthma_crudeprev', 'chd_crudeprev', 'copd_crudeprev']
```

```python
cols_to_drop = [col for col in cdc_df.columns if col not in cols_to_keep]
```

```python
lac_cdc_df = cdc_df.drop(cols_to_drop, axis=1)
lac_cdc_df.columns = ['Blood Pressure', 'Cancer', 'Asthma', 'CHD', 'COPD', 'Geolocation', 'Census tract']
lac_cdc_df['Census tract'] = lac_cdc_df['Census tract'].apply(str).str[1:]
lac_cdc_df[['Blood Pressure', 'Cancer', 'Asthma', 'CHD', 'COPD']] = lac_cdc_df[['Blood Pressure', 'Cancer', 'Asthma', 'CHD', 'COPD']].astype(float)
```

```python
lac_cdc_df = lac_cdc_df.drop_duplicates(subset=['Census tract'], keep='first')
lac_cdc_df
```

```python
import requests
```

```python
endpoint = "https://www.purpleair.com/json"
```

```python
response = requests.get(endpoint).json() # get all sensor data
```

```python
response
```

```python
purple_air_df = pd.DataFrame.from_records(response['results'])
```

```python
# ROUGH boundaries of LA County
north_west = {'Lat': 34.814597, 'Lon': -118.938110}
south_west = {'Lat': 33.724877, 'Lon': -118.886869}
north_east = {'Lat': 34.822724, 'Lon': -117.658021}
south_east = {'Lat': 33.698211, 'Lon': -117.690078}

geo_fcc_endpoint = 'https://geo.fcc.gov/api/census/block/find'


purple_air_df = purple_air_df[:2000]

geo_results = []
for index, row in purple_air_df.iterrows():
    lat = row['Lat']
    lon = row['Lon']
    if(lat != None and lon != None):
        if(lat <= north_west.get('Lat') and lon >= north_west.get('Lon') 
           and lat <= north_east.get('Lat') and lon <= north_east.get('Lon')
           and lat >= south_west.get('Lat') and lon >= south_west.get('Lon')
           and lat >= south_east.get('Lat') and lon <= south_east.get('Lon')):
            
            geo_result = requests.get(f'{geo_fcc_endpoint}?latitude={lat}&longitude={lon}&showall=true&format=json').json()
            
            if(geo_result['status'] != "OK"):
                purple_air_df = purple_air_df.drop(axis=1, index=index)
            elif(geo_result['County']['FIPS'] == '06037'):
                geo_results.append(geo_result['Block']['FIPS'][1:11])   
            else:
                purple_air_df = purple_air_df.drop(axis=1, index=index)
        else:
            purple_air_df = purple_air_df.drop(axis=1, index=index)
    else:
        purple_air_df = purple_air_df.drop(axis=1, index=index)

purple_air_df['Census tract'] = geo_results      
```

```python
print(len(purple_air_df))
print(len(geo_results))
```

```python
purple_air_cols = ['Census tract', 'Lat', 'Lon', 'PM2_5Value']
purple_air_df['PM2_5Value'] = purple_air_df['PM2_5Value'].astype(float)
purple_air_df[purple_air_cols]
```

```python
combined_data_df = pd.merge(lac_cdc_df[['Census tract', 'Blood Pressure', 'Cancer', 'Asthma', 'CHD', 'COPD']], purple_air_df[['Census tract', 'PM2_5Value']])
combined_data_df
```

```python
correlation = combined_data_df.corr().abs()
correlation
```

```python
correlation = correlation.drop(correlation.index[[5]])
correlation = correlation.drop(correlation.columns[[0,1,2,3,4]], axis=1)
correlation
```
