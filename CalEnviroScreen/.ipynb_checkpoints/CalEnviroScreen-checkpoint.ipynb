```python
import pandas as pd
import numpy as np
import missingno as msno
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('ces3results.xlsx')
```

```python
lac_df = df.loc[df['California County'] == 'Los Angeles']
```

```python
lac_df.columns
```

```python
pctl_cols = [col for col in lac_df.columns if 'Pctl' in col or 'Percentile' in col]
score_cols = [col for col in lac_df.columns if 'Score' in col]
drop_cols = pctl_cols + score_cols + ['ZIP', 'Nearby City \n(to help approximate location only)', 'Longitude', 'Latitude', 'SB 535 Disadvantaged Community', 'Housing Burden', 'Unemployment', 'Poverty', 'Linguistic Isolation', 'Education']
print(drop_cols)
```

```python
lac_df_dropped = lac_df.drop(drop_cols, axis=1)
```

```python
lac_df_dropped.columns
```

```python
labels = ['Asthma', 'Low Birth Weight', 'Cardiovascular Disease']
indicators = ['Ozone', 'PM2.5', 'Diesel PM', 'Drinking Water', 'Pesticides', 'Tox. Release',
       'Traffic', 'Cleanup Sites', 'Groundwater Threats', 'Haz. Waste',
       'Imp. Water Bodies', 'Solid Waste']
```

```python
msno.matrix(lac_df_dropped)
```

```python
sum(lac_df_dropped.isnull().sum(axis=1)>3)
```

```python
corr = lac_df_dropped.corr()
```

```python
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns)
```

```python
print(corr.abs().unstack().sort_values(ascending=False))
```

```python
pairs_to_drop = set()
cols = lac_df_dropped.columns
for i in range(0, lac_df_dropped.shape[1]):
    for j in range(0, i+1):
        pairs_to_drop.add((cols[i], cols[j]))
```

```python
au_corr = lac_df_dropped.corr().abs()
```

```python
au_corr
```

```python
cor_target = au_corr[labels]

relevant_features = cor_target.loc[indicators]
relevant_features
```

```python
sns.heatmap(relevant_features)
```
